```python
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
import seaborn as sn
import scipy.stats as stats
# Given data
companies = ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J.P.Morgan & Co.',
             'Lehman Brothers', 'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley',
             'Sun Microsystems', 'Travelers', 'US Airways', 'Warner-Lambert']
measure_X = [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(companies, measure_X, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Measure X (%)')
plt.xticks(rotation=90)
plt.title('Measure X for Different Companies')
plt.tight_layout()
plt.show()
```

```python
# Given data
measure_X = [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00]

# Calculate mean
mean_X = np.mean(measure_X)

# Calculate standard deviation
std_dev_X = np.std(measure_X)

# Calculate variance
variance_X = np.var(measure_X)

print(f"Mean (μ): {mean_X:.2f}")
print(f"Standard Deviation (σ): {std_dev_X:.2f}")
print(f"Variance (σ^2): {variance_X:.2f}")

```

```python
import numpy as np
returns = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]
# Calculate the expected return
expected_return = np.dot(returns, probabilities)

# Calculate the standard deviation
variance = np.dot((np.array(returns) - expected_return)**2, probabilities)
std_dev = np.sqrt(variance)

print("Expected Return:", expected_return)
print("Standard Deviation:", std_dev)
```
